#### seaborn 스타일링 

- set_style: background color, grid, spine, tick을 정의하여 그림의 전반적인 모양을 스타일링
- set_context: 프리젠테이션이나 보고서와 같은 다양한 매체에 활용할 수 있도록 스타일링

1. set_style: 그림의 전반적인 모양 스타일링
    - Built-in Themes (내장 테마) 활용하기
        - Seaborn에는 5가지 기본 제공 테마가 있음 
        - darkgrid, whitegrid, dark, white, ticks. 
        - 기본값은 darkgrid이지만, 원하는대로 변경이 가능
2. sns.despin() 함수
    - 축선 표시 여부 결정하는 함수(right=T/F, left=T/F, top=T/F, bottom=T/F)
        - 각 속성값이 
        - True 면 : 표시하지 않는다.
        - False 면: 표시함
3. sns.set_context() 함수    
    - 전체 스케일 조정
    - paper, notebook, talk, poster. 여기서 기본값은 notebook이다.
    - 스케일로도 전반적인 사이즈를 조정할 수 있지만 실제 폰트 크기를 조정하는 파라미터는 별도로 있음

    - sns.set_context() 안에 font_scale 사용

```python
#seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

                      # whitegrid 는 grid 선을 추가해줌
sns.set_style('dark') # 배경 회색
plt.figure()
plt.plot(x, y1, x, y2, x, y3, x, y4)
```

```python
#seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

sns.set_style('whitegrid') # whitegrid 는 grid 선을 추가해줌
plt.figure()
plt.plot(x, y1, x, y2, x, y3, x, y4)

sns.despine()
```

```python
#sns.despin() 함수
#축선 표시 여부 결정하는 함수(right=T/F,left=T/F,top=T/F,bottom=T/F)
#각 속성값이 True 면 : 표시하지 않는다.
#            False 면: 표시함
x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

sns.set_style('whitegrid') # whitegrid 는 grid 선을 추가해줌
plt.figure()
plt.plot(x, y1, x, y2, x, y3, x, y4)

sns.despine(right = True, left = False, top = True, bottom = False)
plt.show()
```

```python
#seaborn이용한 크기 설정

x=np.linspace(1, 14, 100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

# paper, notebook, talk, poster.
plt.figure(figsize=(10,6))
sns.set_context('talk') # 정해져 있는 크기로 설정해 줄 수 있다.
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
#seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

# paper, notebook, talk, poster.
plt.figure(figsize=(10,6))
sns.set_context('paper', font_scale = .5) # 폰트 크기도 조절해줄 수 있다.
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

###  Distribution Plot
- Distribution Plot은 데이터의 분포를 시각화하는데 도움이 됨
- 이 그래프를 사용하여 데이터의 평균(mean), 중위수(median), 범위(range), 분산(variance), 편차(deviation) 등을 이해 할 수 있음

1. Hist Plot
- 변수에 대한 히스토그램을 표시한다. 막대그래프로 나타난다.
- 하나 혹은 두 개의 변수 분포를 나타내는 전형적인 시각화 도구로 범위에 포함화는 관측수를 세어 표시한다.
- 연속 data 를 확인할 때 사용

```python
sns.histplot(x=tips['total_bill'])
```

2. KDE Plot
- 하나 혹은 두 개의 변수에 대한 분포를 그림
- histplot은 절대량이라면 kdeplot은 밀도 추정치를 시각화
    - 결과물로는 연속된 곡선의 그래프를 얻을 수 있음, 연속된 값을 쭉 연결해서 표현

```python
sns.kdeplot(x=tips['total_bill'])
```

3. Box Plot
- 최대(maximum), 최소(minimum), mean(평균), 1 사분위수(first quartile), 3 사분위수(third quartile)를 보기 위한 그래프
- 특이치(outliar)를 발견하기에도 좋다.
- 단일 연속형 변수에 대해 수치를 표시하거나, 연속형 변수를 기반으로 서로 다른 범주형 변수를 분석할 수 있다.

```python
#tips 데이터를 이용한 box plot 그리기
#seaborn 패키지의 boxplot() 이용해서 그리기

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(x = tips['total_bill'])
sns.despine()
plt.show()
```

```python
#tips 데이터를 이용한 box plot 그리기
#seaborn 패키지의 boxplot() 이용해서 그리기
#요일별 영수증 금액의 분포 확인

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(x = tips['day'], y = tips['total_bill'])
sns.despine(offset = 10) # x, y 축간의 간격을 설정하는 인수
plt.show()
```

```python
#tips 데이터를 이용한 box plot 그리기
#seaborn 패키지의 boxplot() 이용해서 그리기
#요일별 영수증 금액의 분포 확인
#palette 인수 사용 예제 : palette = '파레트명'

sns.set_style('whitegrid')

plt.figure(figsize=(10, 6))
sns.boxplot(x = 'day', y ='total_bill', data = tips, palette='Set2')
sns.despine(offset = 10)
plt.show()
```

```python
# hue 인수 : 카테고리값을 가지는 변수의 이름 지정하면
# 카테고리에 따라 그룹을 만들어서 표시
# 요일별 흡연여부에 따른 음식값의 분포
plt.figure(figsize=(10,6))
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips, palette = 'Set3')
plt.show()
```
