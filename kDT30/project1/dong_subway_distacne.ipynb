## 다세대 주택별로 가까운 지하철역 특정(위치 특성과 관련)

```python
# index 를 기준으로 거리행렬을 만든다.
# 연립다세대, 역명을 index로 설정
df.set_index(['연립다세대'], inplace = True)
df3.set_index(['역명'], inplace = True)
```

```python
df.head(1)
df3.head(1)
```

```python
# 거리 행렬 생성을 위한 컬럼을 추출 후 df 로 생성
# 먼저 나타나는 역이 더 가까운 역이다.
df_location = df[['위도', '경도']]
df3_location = df3[['위도', '경도']]
```

```python
# 결측치가 있는지를 확인
# 계산이 가능한 정수, 실수형인지를 확인한다.
df_location.info()
df3_location.info()
```

```python
df_location.head(5)
df3_location.head(5)
```

```python
# !pip install haversine
# 거리 계산을 위함
```

```python
# 거리행렬 생성
from scipy.spatial.distance import cdist
from haversine import haversine

# df_locaiton(아파트) - 행, df3.location(지하철) - 열 앞쪽이 행, 뒤쪽이 열로 구성이 된다.
dist_mat = cdist(df_location, df3_location, metric = haversine)
```

```python
# 계산 값이 행렬 array로 반환이 된다.
dist_mat
```

```python
# pd.DataFrame(dist_mat)
# 682번 지역에서 가장 가까운 지하철역까지의 거리를 구했다.
pd.DataFrame(dist_mat).loc[682].min()
```

```python
# index 번호 => 지하철역명
df_test = pd.DataFrame(dist_mat, columns = df3.index, index = df.index)
df_test.head()
```

```python
df_test.to_csv('C:/Users/gherwt/KDT30/p1/dist.csv')
```

```python
df_arr = df_test.to_numpy() # np의 array 자료구조로 변환시켜주는 함수
dist_mat # 위에서부터 진행되어 왔기 때문에 array 가 저장되어 있다.
```

```python
close_sub_index = df_arr.argsort(axis = 1)
pd.DataFrame(close_sub_index) # 전체행렬임, 0열에 가장 가까운 지하철역의 index가 존재 -> 행을 기준으로 오름차순정렬했기 때문
```

```python
# pd.DataFrame(close_sub_index)[[0]] index 가 0인 값만 모아준 행렬
close_sub_index = df_arr.argsort(axis = 1)[:,0]
close_sub_index # 모든 연립주택의 가장 가까운 역의 지하철역의 index만 모아서 반환
```

```python
df3.iloc[close_sub_index].index # 연립주택에서 가장 가까운 지하철역 명, 인덱스 개체
df3.iloc[close_sub_index].index.values # array 형태로 밀어서 만들어줘야 한다. value 만 추출해서 만들어주면 된다.
```

```python
# 연립주택 및 주소 정보 df
df['가까운역'] = df3.iloc[close_sub_index].index.values
df.head(2)
```

```python
# 가장 가까운 역과의 거리
df['가까운역과의 거리'] = pd.DataFrame(dist_mat).min(axis = 1).values
df.head(2)
```
