```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import os
import pandas as pd
import numpy as np
import geopy.distance
```

```python
# 성동구 데이터 불러오기
df1 = pd.read_csv("C:/Users/gherwt/KDT30/p1/dong.csv", encoding = "cp949")
df2 = pd.read_csv("C:/Users/gherwt/KDT30/p1/school_location.csv", encoding = "cp949")
```

```python
# 위도 경도 이름 변경
df1 = df1.rename(columns={'위도':'dong_lat'})
df1 = df1.rename(columns={'경도':'dong_long'})
```

```python
# 위도 경도 이름 변경
df2 = df2.rename(columns={'위도':'school_lat'})
df2 = df2.rename(columns={'경도':'school_long'})
```

```python
# 법정동을 기준으로 병합해준다.
data = df2.merge(df1 , on = '법정동' , how = 'left') # 데이터 합치기
data.head(2)
```

```python
# zip 함수로 위경도 묶어주기
data['dong_coord'] = pd.Series(zip(data['dong_lat'], data['dong_long']))
data['school_coord'] = pd.Series(zip(data['school_lat'], data['school_long']))
data.head(2)
```

```python
# 위, 경도 값의 차를 이용해서 거리를 구해준다.
data['distance'] = data.apply(lambda x : geopy.distance.distance(x['dong_coord'] , x['school_coord']).km , axis =1 ) # 거리계산
```

```python
data.head(2)
```

```python
# 거리 df 를 만들어준다.
distance = data[['학교명', '주소', '법정동', 'distance']]
```

```python
distance
```

```python
# 동을 기준으로 묶고 그 동에 있는 학교의 거리 평균을 구해준다.
distance_mean = distance.groupby('법정동')['distance'].mean()
```
