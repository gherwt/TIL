```python
# geopy 를 사용해 주소로 위치 좌표 구하기
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
# 미리 처리해 놓은 학교 주소 data 를 불러와준다.
education = pd.read_csv('C:/Users/gherwt/KDT30/p1/educations.csv', encoding = "cp949")
address = education['도로명주소']
address.head()
```

```python
# 분리되어 있는 주소를 띄어쓰기로 합쳐준다.
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
print(address)
```

```python
# local -> 한국을 기준으로 설정한다.
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        # geo_local을 활용 address 에 있는 주소 로 위, 경도를 추출해준다.
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
# 주소를 위,경도 값으로 변환하기
latitude = []
longitude =[]

# 비어있는 list 에 추가해준다.
for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
# 만들어진 위, 경도로 학교의 위치 좌표를 포함한 df 를 만들어준다.
address_df = pd.DataFrame({'학교명': education['학교명'],'주소':education['도로명주소'],'법정동':education['도로명상세주소'],'주소':address,'위도':latitude,'경도':longitude})
```

```python
address_df.to_csv('school_location.csv')
```
