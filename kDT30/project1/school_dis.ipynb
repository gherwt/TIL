```python
# 각 동에서 최소 거리에 있는 학교값을 추출한다.
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import os
import pandas as pd
import numpy as np
import geopy.distance
import reverse_geocode
from scipy.spatial.distance import cdist
from haversine import haversine
```

```python
# 성동구 데이터 불러오기
df1 = pd.read_csv("C:/Users/gherwt/KDT30/p1/dong.csv", encoding = "cp949")
df2 = pd.read_csv("C:/Users/gherwt/KDT30/p1/school_location.csv", encoding = "cp949")
```

```python
# 위도 경도 이름 변경
df1 = df1.rename(columns={'위도':'dong_lat'})
df1 = df1.rename(columns={'경도':'dong_long'})
```

```python
# 위도 경도 이름 변경
df2 = df2.rename(columns={'위도':'school_lat'})
df2 = df2.rename(columns={'경도':'school_long'})
```

```python
df2.head(2)
```

```python
# 각 df 의 위도, 경도 data 를 따로 저장해준다.
df1_location = df1[['dong_lat', 'dong_long']]
df2_location = df2[['school_lat', 'school_long']]
```

```python
# cdist 를 사용해서 동과 각 학교 사이의 거리를 구해준다.
dist_mat = cdist(df1_location, df2_location, metric = haversine)
df_test = pd.DataFrame(dist_mat, columns = df2.index, index = df1.index)
df_test.head()
```

```python
# np의 array 자료구조로 변환
df_arr = df_test.to_numpy()

# 0열에 가장 가까운 학교와의 거리에 대한 index가 존재 -> 오름차순정렬함
close_sub_index = df_arr.argsort(axis = 1)
pd.DataFrame(close_sub_index)
```

```python
# pd.DataFrame(close_sub_index)[[0]] index 가 0인 값만 모아준 행렬
close_sub_index = df_arr.argsort(axis = 1)[:,0]
```

```python
# 동에서 가장 가까운 학교와의 거리
df1['school_dis'] = pd.DataFrame(dist_mat).min(axis = 1).values
```
